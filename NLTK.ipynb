```python
import nltk
nltk.download('punkt', quiet=False)  # Ensure punkt tokenizer is downloaded
nltk.download('punkt_tab', quiet=False)  # Download punkt_tab if needed
from nltk.tokenize import word_tokenize
text ="NLTK seminar in technogis is fun"
tokens = word_tokenize(text)
print(tokens)
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
text = "Spacy seminar in technogis is fun"
doc = nlp(text)

for ent in doc.ents:
    print(ent.text, ent.label_)
```
